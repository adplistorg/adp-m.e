```python
r = requests.get("https://api.adplist.org/mentor/?on_break=false&offset=0&limit=5000&")
r.text
```

```python
y = json.loads(r.text)
```

```python
x = requests.get("https://api.adplist.org/group-session/?limit=50000&offset=0")
x.text
```

```python
a = json.loads(x.text)
```

```python
a['results'][0]
```

```python
for i in range(len(a['results'])):
    if type(a['results'][i]['mentor']) == list:
        print(a['results'][i]['name'])
```

```python
type(a['results']) == list
```

```python
df = pd.DataFrame(columns = ['Mentor', 'Reviews', 'Sessions'])
```

```python
for i in range(len(y['results'])):
    df.loc[i,'Mentor'] = y['results'][i]['name']
    df.loc[i,'Reviews'] = y['results'][i]['total_reviews']
    df.loc[i, 'Sessions'] = 0
```

```python
df
```

```python
for i in range(len(df)):
    count = 0
    mentor_name = df.loc[i,'Mentor']
    for j in range(len(a['results'])):
        if a['results'][j]['mentor']['name'] == mentor_name:
            count+=1
    if count>0:
        df.loc[i,'Sessions'] = count
        
```

```python
df_final = pd.DataFrame(columns = ['Mentor'])
```

```python
n = 0
for i in range(len(df)):
    if df.loc[i,'Reviews'] == 0 and df.loc[i, 'Sessions'] == 0:
        df_final.loc[n,'Mentor'] = df.loc[i,'Mentor']
        n+=1
```
